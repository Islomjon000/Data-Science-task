# Birinchi model (rasm klassifikatsiyasi)

```python
from os import path
from fastai.vision.all import *
from ipywidgets import widgets

# data
path = untar_data(URLs.PETS)/'images'

# labellovchi funksiya
def is_cat(x): return x[0].isupper()

#datalarni yuklovchi (dataloaders)
dls = ImageDataLoaders.from_name_func(
    path, get_image_files(path), valid_pct=0.2, seed=42,
    label_func=is_cat, item_tfms=Resize(224)
)

# o'qitish train
learn = cnn_learner(dls, resnet34, metrics=accuracy)
learn.fine_tune(2)

```

```python
# rasm yuklab olish uchun widget
upload = widgets.FileUpload()
upload
```

```python

img = PILImage.create(upload.data[-1]) #rasm
pred, _, prob = learn.predict(img) #tahmin
print(f"Mushukmi?: {pred}")
print(f"Ehtimollik: {prob[1].item():.3f}")
img
```
